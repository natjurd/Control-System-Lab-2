```python
!pip install control
import control as C #import control systems lib
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
```

```python

M = 0.3
m = 0.1
ell = 0.35
g = 9.81

a = 4 / ((4*M) + m)
b = (3*m*g) / ((4*M) + m)
c = 3 / (ell*(4*M + m))
d = (3*g*(M + m)) / (ell*((4*M) + m))
s = C.TransferFunction.s

G_theta = -c / (s**2 - d)
G_x = (a*(s**2 - d) + b*c) / (s**2*(s**2 - d))
n_point = 1000
t_final = 0.2
t_span = np.linspace (0, t_final, n_point)
ip_signal = np.sin(100*t_span**2)

result1 = C.forced_response (G_theta, t_span, ip_signal)
plt.xlabel('Timespan,t(s)')
plt.ylabel('Angle of rod(\u03B8)')
plt.plot (result1.time, result1.outputs, '-r')
plt.grid()
plt.show()

result2 = C.forced_response (G_x, t_span, ip_signal)
plt.xlabel('Timespan,t(s)')
plt.ylabel('Horizontal position of car,x(m)')
plt.plot (result2.time, result2.outputs, '-k')
plt.grid()
plt.show()
```
